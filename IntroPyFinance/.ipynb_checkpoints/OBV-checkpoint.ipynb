```python
from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf

raw = yf.download('SPY AAPL', start = '2010-01-01', end = '2022-12-31')
raw

def fix_cols(df):
    columns = df.columns
    outer = [col[0] for col in columns]
    df.columns = outer
    return df

(raw
 .iloc[:, ::2]
 .pipe(fix_cols)
)

def tweak_data():
    raw = yf.download('SPY AAPL', start = '2010-01-01', end = '2019-12-31')
                    
    return(raw
     .iloc[:, :-2:2] # :-2:2 takes out volume, ::2 leaves volume in
     .pipe(fix_cols)
)
                      
tweak_data()


# TO define apple data
aapl = (raw
        .iloc[:,::2]
        .pipe(fix_cols))
aapl
```

```python
# To refactor

def calc_obv(df, close_col='Close', vol_col='Volume'):
    close = df[close_col]
    vol = df[vol_col]
    close_shift = close.shift(1)
    return (df
            .assign(vol=np.select([close > close_shift,
                                   close == close_shift,
                                   close < close_shift],
                                  [vol, 0, -vol]),
                    obv=lambda df_:df_.vol.fillna(0).cumsum()
                   )
            ['obv']
           )

(aapl
 .assign(obv = calc_obv)
)
```

```python
# How it can be tested with values from excel sheet, small dataset, etc

test_df = pd.DataFrame({'Close': [....],
                        'Volume': [....]})
(test_df
 .assign(obv=calc_obv)
)
```
