```python
from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf

raw = yf.download('SPY AAPL', start = '2010-01-01', end = '2019-12-31')
raw

def fix_cols(df):
    columns = df.columns
    outer = [col[0] for col in columns]
    df.columns = outer
    return df

(raw
 .iloc[:, ::2]
 .pipe(fix_cols)
)

import yfinance as yf

def tweak_data():
    raw = yf.download('SPY AAPL', start = '2010-01-01', end = '2019-12-31')
    
    return(raw
     .iloc[:, :-2:2]
     .pipe(fix_cols)
          )

aapl = (raw
        .iloc[:,::2]
        .pipe(fix_cols))
```

## Plotting Returns

```python
# Plot Returns

(aapl
 .pct_change()
 .Close
 .plot()
)
```

```python
# Histogram of Returns
(aapl
 .pct_change()
 .Close
 .hist()
)
```

```python
# Change bins
(aapl
 .pct_change()
 .Close
 .hist(bins=250)
)
```

```python
# Understanding plotting in Pandas
# is a huge lever in Bar Plot Returns
(aapl
 .pct_change()
 .Close
 .iloc[-100:] # lets take our last 100 values of our "Close" data # No comma in .iloc -> no column selection
 .plot.bar()
)
```

```python
# Bar Plot of Returns
# Sadly dates are broken with Pandas bar plots
# 1970s?
fic, ax = plt.subplots(figsize=(10,4))
(aapl
 .pct_change()
 .Close
 .iloc[-100:]
 .plot.bar(ax=ax)
)

ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
ax.xaxis.set_minor_locator(dates.DayLocator())

# this plot will not have correct dates. because pandas has converted them to categorical values
```

```python
# Returns using matplotlib

def my_bar(ser, ax):
    ax.bar(ser.index, ser)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return ser



fig,ax = plt.subplots(figsize=(10,4))
(aapl
 .pct_change()
 .Close
 .iloc[-100:]
 .pipe(my_bar, ax)
)

```
