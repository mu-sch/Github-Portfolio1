## Plot the rolling volatility over 30-day sliding windows for 2015-2022 for Amazon(AMZN)

```python
from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf

raw = yf.download('SPY AMZN', start = '2010-01-01', end = '2022-12-31')
raw

def fix_cols(df):
    columns = df.columns
    outer = [col[0] for col in columns] # using a list comprehension to extract the first entry for each tuple in 
                                        # the multi-index from .columns ^above^
    df.columns = outer
    return df

(raw
 .iloc[:, ::2]
 .pipe(fix_cols)
)

import yfinance as yf
def fix_cols(df):
    columns = df.columns
    outer = [col[0] for col in columns] 
    df.columns = outer
    return df

def tweak_data():
    raw = yf.download('SPY AMZN', start = '2010-01-01', end = '2022-12-31')
                    
    return(raw
     .iloc[:, :-2:2] # :-2:2 takes out volume, ::2 leaves volume in
     .pipe(fix_cols)
)
                      
tweak_data()

amzn = (raw
        .iloc[:,::2]
        .pipe(fix_cols))
```

```python
(amzn
 .Close
 .mean()
)
```

```python
(amzn
 .assign(pct_change=amzn.Close.pct_change())
 .rolling(30, min_periods=30)
 .pct_change
 .std()
 .loc['2015' : '2022']
 .plot()
)
```
