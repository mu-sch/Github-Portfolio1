```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import pandas as pd
import numpy as np
from scipy.optimize import minimize 
import edhec_risk_kit as erk

ind = erk.get_ind_returns()
er = erk.annualize_rets(ind["1996":"2000"], 12)
cov = ind["1996":"2000"].cov()
```

```python
l = ["Food", "Steel"]

# To compute max sharpe ratio. Will return the respective weights of the assets for max sharpe ratio
erk.msr(0.1, er[l], cov.loc[l,l])
```
